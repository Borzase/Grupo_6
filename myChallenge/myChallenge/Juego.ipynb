```python
import numpy as np
import random

class Barco:
    def __init__(self, tamano):
        self.tamano = tamano

class Jugador:
    def __init__(self, nombre, es_maquina=False):
        self.nombre = nombre
        self.es_maquina = es_maquina
        self.tablero_j = np.full((10, 10), ' ')  # Tablero donde coloca sus barcos
        self.tablero_disparos = np.full((10, 10), ' ')  # Tablero donde registra sus disparos
        self.barcos = self.crear_barcos()
    
    def crear_barcos(self):
        barcos = []
        barcos += [Barco(1) for _ in range(4)]  # 4 barcos de tamaño 1
        barcos += [Barco(2) for _ in range(3)]  # 3 barcos de tamaño 2
        barcos += [Barco(3) for _ in range(2)]  # 2 barcos de tamaño 3
        barcos.append(Barco(4))  # 1 barco de tamaño 4
        return barcos
    
    def colocar_barcos(self):
        for barco in self.barcos:
            while True:
                if self.es_maquina:
                    x, y = random.randint(0, 9), random.randint(0, 9)
                    direccion = random.choice(['N', 'S', 'E', 'O'])
                else:
                    x = int(input(f"{self.nombre}, ingresa la coordenada X para tu barco de tamaño {barco.tamano}: "))
                    y = int(input(f"{self.nombre}, ingresa la coordenada Y para tu barco de tamaño {barco.tamano}: "))
                    direccion = input("Ingresa la dirección (N/S/E/O): ")
                
                if colocar_barco_en_tablero(self.tablero_j, barco, x, y, direccion):
                    break
    
    def disparar(self, oponente):
        if self.es_maquina:
            x, y = random.randint(0, 9), random.randint(0, 9)
        else:
            x = int(input(f"{self.nombre}, ingresa la coordenada X para disparar: "))
            y = int(input(f"{self.nombre}, ingresa la coordenada Y para disparar: "))
        
        if oponente.tablero_j[y, x] == 'O':
            print("¡Impacto!")
            oponente.tablero_j[y, x] = 'X'
            self.tablero_disparos[y, x] = 'X'
        else:
            print("Agua...")
            self.tablero_disparos[y, x] = '-'
    
    def mostrar_tableros(self):
        print(f"Tablero de {self.nombre} (Posición de barcos):")
        print(self.tablero_j)
        print(f"Tablero de {self.nombre} (Disparos realizados):")
        print(self.tablero_disparos)

def colocar_barco_en_tablero(tablero, barco, x, y, direccion):
    if direccion == 'N' and y - (barco.tamano - 1) < 0:
        return False
    if direccion == 'S' and y + (barco.tamano - 1) >= 10:
        return False
    if direccion == 'E' and x + (barco.tamano - 1) >= 10:
        return False
    if direccion == 'O' and x - (barco.tamano - 1) < 0:
        return False
    
    for i in range(barco.tamano):
        if direccion == 'N' and tablero[y - i, x] == 'O':
            return False
        if direccion == 'S' and tablero[y + i, x] == 'O':
            return False
        if direccion == 'E' and tablero[y, x + i] == 'O':
            return False
        if direccion == 'O' and tablero[y, x - i] == 'O':
            return False
    
    for i in range(barco.tamano):
        if direccion == 'N':
            tablero[y - i, x] = 'O'
        if direccion == 'S':
            tablero[y + i, x] = 'O'
        if direccion == 'E':
            tablero[y, x + i] = 'O'
        if direccion == 'O':
            tablero[y, x - i] = 'O'
    
    return True

def jugar():
    jugador = Jugador("Jugador", es_maquina=False)
    maquina = Jugador("Máquina", es_maquina=True)
    
    print("Colocación de barcos:")
    jugador.colocar_barcos()
    maquina.colocar_barcos()
    
    turno = 0
    while True:
        print("\nEstado actual del juego:")
        jugador.mostrar_tableros()
        maquina.mostrar_tableros()
        
        if turno % 2 == 0:
            print("Turno del jugador")
            jugador.disparar(maquina)
        else:
            print("Turno de la máquina")
            maquina.disparar(jugador)
        
        turno += 1
        
        if np.all(maquina.tablero_j != 'O'):
            print("¡Ganaste!")
            break
        if np.all(jugador.tablero_j != 'O'):
            print("La máquina ha ganado.")

# Aquí es donde se llama a la función 'jugar'

jugar()
```
